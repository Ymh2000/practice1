```python
import pandas as pd
import cv2

# 1. CSV 읽기
df = pd.read_csv('work_list.csv', encoding='utf-8-sig')

# 2. '작업자'가 '윤명현'인 행 필터링
df = df[df['작업자'] == '윤명현']

jpg_list = df[df['jpg_path'].str.endswith('.jpg', na=False)]

# 4. 각 이미지를 열어서 OpenCV로 표시
for path in jpg_list['jpg_path']:
    image = cv2.imread(path)
    if image is not None:
        cv2.imshow("Image", image)
        cv2.waitKey(0)  # 키 입력 대기
        cv2.destroyAllWindows()
    else:
        print(f"이미지를 열 수 없습니다: {path}")
```
